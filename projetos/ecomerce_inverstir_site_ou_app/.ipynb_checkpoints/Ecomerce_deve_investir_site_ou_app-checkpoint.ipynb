# Agora vamos para a construo do Modelo

## Treinando e testando os dados

O primeiro passo  dividir os dados em conjuntos de treinamento e teste.

** Definindo as uma variveis: X = todas as caractersticas numricas dos clientes e Y = Valor anual gasto (Yearly Amount Spent). **

```python
y = clientes['Yearly Amount Spent']
```

```python
X= clientes[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
# Importando scikit-learn para dividir os dados em conjunto de treinamento e conjunto de teste.
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test =train_test_split(X, y, test_size=0.3,random_state=101)
```

## Treinando o modelo

Agora  hora de treinar nosso modelo em nossos dados de treinamento!

```python
# Importando o LinearRegression do sklearn.linear_model
from sklearn.linear_model import LinearRegression
```

** Criando uma instncia de um modelo LinearRegression () chamado lm. **

```python
lm = LinearRegression()
```

** Treinando lm nos dados de treinamento. **

```python
lm.fit(X_train, y_train)
```

**Resultado dos coeficientes do modelo**

```python
lm.coef_
```

## Previso de dados de teste
Agora que nos ajustamos ao nosso modelo, vamos avaliar o seu desempenho ao prever os valores de teste!


```python
# Usando o lm.predict() para prever o conjunto X_test dos dados.
prediction = lm.predict(X_test)
```

```python
# Agora criamos um diagrama de disperso (scatterplot) dos valores reais de teste em relao aos valores preditos 
# para ver se o modelo est fazendo uma boa predio.

plt.scatter(y_test, prediction)
plt.xlabel('Y test')
plt.ylabel('Predicted Y')
```

** Como podemos ver os valores preditos e os de teste esto bem relacionados, mostrando que estamos no caminho certo com nosso modelo.**

## Avaliando o Modelo

Vamos avaliar o desempenho do nosso modelo calculando a soma residual dos quadrados e o escore de varincia explicado (R ^ 2).

** Calcule o erro absoluto mdio, o erro quadrado mdio e o erro quadrado mdio da raiz. Consulte a palestra ou a Wikipdia para as frmulas **

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```

## Resduos

Pelos resultados podemos ver que obtivemos um modelo muito bom com um bom ajuste. Vamos explorar rapidamente os resduos para garantir que tudo esteja bem com os dados.

```python
# Criando um histograma dos resduos e certificando-se de que ele parece normalmente distribudo. 
sns.distplot((y_test-prediction), bins=50)
```

## Concluso

```python
coef = pd.DataFrame(lm.coef_, X.columns, columns=['Coefs'])
coef
```

**Mantendo todos as outras variveis constantes**, um aumento de 1 unidade no tempo gasto no App est associado a um aumento de **38,59 dlares** totais gastos.

**Mantendo todos as outras variveis constantes**, um aumento de 1 unidade no tempo no site est associado a um aumento de **0,19 dlares** totais gastos.

**Mantendo todos as outras variveis constantes**, um aumento de 1 unidade no tempo de Associao est associado a um aumento de **61,27 dlares** totais gastos.

***Diante dos resultados obtidos, podemos ver que a empresa, primeiramente, deveria arranjar outras formas de fidelizar seu cliente, j que essa  a varivel que mais ifluenciam nos gastos dos seus usurios. Entre site e aplicativo, a empresa deve investir em aplicavo, dado que o mesmo apresenta um coeficiente significativamente maior do que o site. ***
